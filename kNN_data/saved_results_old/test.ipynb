```python
import pandas as pd

filename = 'Results on CBF (0.01 to 0.1)' 
def insert_second_column(file_main, file_extra, output_file):
    # Read the two CSV files (auto-detect delimiter if needed)
    df_main = pd.read_csv(file_main)
    df_extra = pd.read_csv(file_extra)
    
    # Extract the second column from the extra file
    col_to_insert = df_extra.iloc[:, 1]  # index 1 means second column
    col_name = df_extra.columns[1]
    
    # Insert right after the second column of the main file
    insert_position = 2  # index after first two columns
    df_main.insert(insert_position, col_name, col_to_insert)
    
    # Save the result
    df_main.to_csv(output_file, index=False)

# Example usage
insert_second_column(filename + ".csv",filename +  "__test.csv", filename + ".csv")
```

```python
import matplotlib.pyplot as plt

markers = ['o', 's', '^', 'D', 'v', 'P', 'X']
x = range(len(markers))
y = [1] * len(markers)

plt.figure(figsize=(8, 2))
for i, m in enumerate(markers):
    plt.plot(x[i], y[i], marker=m, markersize=12, linestyle='None', label=m)

plt.ylim(0.5, 1.5)
plt.legend(title="Markers", ncol=len(markers))
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Sample identical data
x = [0, 1, 2, 3, 4]
y = np.array([0, 1, 4, 9, 16])

plt.figure(figsize=(6, 4))

# First series
plt.plot(x, y, color='blue', alpha=0.9, linewidth=3, label='Series 1')

# Second series (identical)
plt.plot(x, y + 0.1, color='red', alpha=0.6, linewidth=3, label='Series 2')

plt.title("Transparency helps show overlap")
plt.legend()
plt.grid(True)
plt.show()
```
