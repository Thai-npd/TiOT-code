```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Example dictionary
data = {
    'Group A': np.random.normal(0, 1, 100),
    'Group B': np.random.normal(1, 1, 100),
    'Group C': np.random.normal(2, 1, 100)
}

# Create a list of data values and labels
data_values = list(data.values())
labels = list(data.keys())
colors = ['skyblue', 'salmon', 'lightgreen']  # Customize colors here

# Create box plot
sns.set(style="whitegrid", context="paper")
fig, ax = plt.subplots()
box = ax.boxplot(data_values, patch_artist=True)  # patch_artist=True enables fill color

# Set colors
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# Customize plot
ax.set_xticks(range(1, len(labels) + 1))
ax.set_xticklabels(labels)
ax.set_ylabel('Values')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Line colors
line_colors = [
    '#d73027',  # Crimson Red
    '#fc8d59',  # Burnt Orange
    '#d08c32',  # Golden Brown
    '#8c6bb1',  # Dusty Purple
    '#2171b5',  # Strong Blue
    '#52ae00',  # Fresh Green
    '#00ae4f'   # Emerald Green
]

# Line labels
labels = [
    'Algo 1 (Red)',
    'Algo 2 (Orange)',
    'Algo 3 (Brown)',
    'Algo 4 (Purple)',
    'Algo 5 (Blue)',
    'Algo 6 (Green A)',
    'Algo 7 (Green B)'
]

# Simulate decreasing loss curves
x = np.linspace(0, 1, 300)  # normalized "epochs"
loss_curves = [
    1.4 * np.exp(-4 * x),                   # fast decay
    1.5 / (1 + 10 * x),                     # inverse decay
    1.3 * np.exp(-2 * x) + 0.05 * x,        # smooth with tail
    1.45 * np.exp(-6 * x**0.8),             # nonlinear steep
    1.25 * np.exp(-3 * x) + 0.03*np.sin(10*x), # some wiggle
    1.6 / (1 + 20 * x**1.5),                # faster inverse decay
    1.35 * np.exp(-3 * x**1.2)              # sharp at first, slows later
]

# Plot
plt.figure(figsize=(10, 5))
for y, color, label in zip(loss_curves, line_colors, labels):
    plt.plot(x, y, color=color, linewidth=2, label=label)

# Style
plt.title('Simulated Objective Loss Curves of 7 Algorithms', fontsize=14, weight='bold')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define color groups
color_map = {
    'red': ['#d73027', '#fc8d59', '#d08c32'],    # For 'NPG' group
    'green': ['#52ae00', '#00ae4f'],             # For 'LS' group
    'blue': ['#8c6bb1', '#2171b5']                # Default group
}

# Sample algorithm names
algorithms = [
     'AccNPG', 'BasicNPG',
    'BacktrackingLS', 'StrongLS',
    'Baseline', 'FastNPG','GD'
]

# Step 1: assign group and color
assigned_colors = []
group_labels = []

# Track usage index for each group
color_index = {'red': 0, 'green': 0, 'blue': 0}

for name in algorithms:
    if 'NPG' in name:
        group = 'red'
    elif 'LS' in name:
        group = 'green'
    else:
        group = 'blue'
    
    # Pick next color in group's list (cycle if needed)
    group_colors = color_map[group]
    index = color_index[group] % len(group_colors)
    assigned_colors.append(group_colors[index])
    color_index[group] += 1
    group_labels.append(f"{name} ({group})")

# Step 2: simulate decreasing losses
x = np.linspace(0, 1, 300)
np.random.seed(0)
loss_curves = []

for i in range(len(algorithms)):
    base = 1.3 + 0.2 * np.random.rand()
    rate = 2.5 + 2 * np.random.rand()
    noise = 0.02 * np.random.randn(len(x))
    decay = base * np.exp(-rate * x) + 0.03 * np.sin(10 * x) + noise
    loss_curves.append(decay)

# Step 3: plot
plt.figure(figsize=(10, 5))
for name, y, color in zip(group_labels, loss_curves, assigned_colors):
    plt.plot(x, y, label=name, color=color, linewidth=2)

plt.title("Loss Curves Grouped by Algorithm Type", fontsize=14, weight='bold')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()
```
